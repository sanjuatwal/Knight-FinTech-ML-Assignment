```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
import seaborn as sns
```

```python
df = pd.read_csv("./Knight ML Assignment/Data/train.csv")
```

```python
df.head()
```

## cheapest and mose expensive wine of each country


```python
df.groupby(['country']).price.agg([min, max])
```

## cheapest wine in each point category

```python
df.groupby('points').price.min()
```

## most expensive  wine in each point category

```python
df.groupby('points').price.max()
```

## average wine price in each point category

```python
df.groupby('points').price.mean()
```

## best wines of each province of each country

```python
df.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])[['points', 'price', 'variety', 'winery']]
```
